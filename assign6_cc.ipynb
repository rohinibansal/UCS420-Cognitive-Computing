Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

 Generate x values from -10 to 10 using np.linspace().

 Compute y values for:

o y= M⋅x

2

o y=M⋅sin(x)

 Plot both functions in a single figure:

o Use different colors and line styles.

o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter the value of M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2  
y2 = M * np.sin(x)  
plt.figure(figsize=(8, 5))
plt.plot(x, y1, 'r--', label=r'$y = Mx^2$')
plt.plot(x, y2, 'b-', label=r'$y = M\sin(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Plots of y = Mx² and y = Msin(x) for M = {M}')
plt.legend()
plt.grid(True)
plt.show()
```

Q.2 Create a dataset of five subjects and scores.

 Convert it into a Pandas DataFrame.

 Plot the scores using a Seaborn bar plot with:

o Different colors for each bar.

o Annotations on top of each bar.

o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'Subject': ['Math', 'Physics', 'Chemistry', 'Biology', 'English'],
        'Score': [85, 78, 90, 88, 76]}
df = pd.DataFrame(data)
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, hue='Subject', palette="viridis", legend=True)
for index, row in df.iterrows():
    ax.text(index, row['Score'] + 1, str(row['Score']), ha='center', fontsize=12)
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Scores in Different Subjects")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().

 Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.
    
o Scatter plot with random noise.

 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))  
np.random.seed(roll_number)  

data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(np.cumsum(data), marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Cumulative Sum Line Plot")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumulative Sum")
axs[0, 0].grid(True)

x_values = np.arange(50)
noise = np.random.randn(50) * 0.5  
axs[0, 1].scatter(x_values, data + noise, color='r', alpha=0.7)
axs[0, 1].set_title("Scatter Plot with Random Noise")
axs[0, 1].set_xlabel("Index")
axs[0, 1].set_ylabel("Values")
axs[0, 1].grid(True)

axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link

https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv

Apply ‘seaborn’ library to do the following.

1. Read Total profit of all months and show it using a line plot.
2. 
3. Read all product sales data and show it using a multiline plot.
4. 
5. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv'
data = pd.read_csv(url)

sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()

sales_data = data.melt(id_vars='month_number', 
                       value_vars=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'],
                       var_name='Product', value_name='Sales')

plt.figure(figsize=(12, 8))
sns.lineplot(x='month_number', y='Sales', hue='Product', data=sales_data, marker='o')
plt.title('Monthly Sales of Different Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.show()

attributes = data.columns[1:]

for attribute in attributes:
    plt.figure(figsize=(10, 6))
    sns.barplot(hue='month_number', y=attribute, data=data, palette='viridis')
    plt.title(f'{attribute.replace("_", " ").title()} per Month')
    plt.xlabel('Month Number')
    plt.ylabel(attribute.replace("_", " ").title())
    plt.show()

```
